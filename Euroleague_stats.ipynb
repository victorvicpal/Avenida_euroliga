```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
#Descarga de Tablas
stats_gen=pd.read_csv('/Users/victor/Documents/DataLab/PERFUMERIAS/teams_stats.csv')
stats_blk=pd.read_csv('/Users/victor/Documents/DataLab/PERFUMERIAS/teams_blk.csv')
stats_reb=pd.read_csv('/Users/victor/Documents/DataLab/PERFUMERIAS/teams_rebounds.csv')
stats_stl=pd.read_csv('/Users/victor/Documents/DataLab/PERFUMERIAS/teams_stl.csv')
stats_ast=pd.read_csv('/Users/victor/Documents/DataLab/PERFUMERIAS/tems_ast.csv')
stats_ft=pd.read_csv('/Users/victor/Documents/DataLab/PERFUMERIAS/teams_ft.csv')
```

```python
#Unión de Tablas
stats=stats_gen.join(stats_blk.set_index('Team'),on='Team')
stats=stats.join(stats_ast.set_index('Team'),on='Team')
stats=stats.join(stats_reb.set_index('Team'),on='Team')
stats=stats.join(stats_stl.set_index('Team'),on='Team')
stats=stats.drop(['BLKPG','TOPG','BLKP40M','APG','TOPG','AP40M','ORPG','DRPG','RPG','STLPG','STLP40M'],axis=1)
```

## Efectividad Anotadora
<img style="float: right;" src="img/silvia.jpg",width="400">
###### En este gráfico podéis comparar las estadísticas de cada equipo en función del porcentage de tiros anotado, el porcentaje de triples convertidos y los puntos encestados.

* x <- FG%
* y <- 3P%
* z <- PTS

Los gráficos han sido realizados enteramente con [Python](https://www.python.org/)+[Plotly](https://plot.ly/)

```python
trace1 = go.Scatter3d(
    x = stats['FG%'],
    y = stats['3P%'],
    z = stats['PTS'],
    mode='markers',
    text=stats['Team'],
    marker=dict(
        size=12,
        color=stats['PTS'],                # set color to an array/list of desired values
        colorscale='Viridis',              # choose a colorscale
        opacity=0.8
    )
)

data = [trace1]

layout = go.Layout(
    title='Efectividad Anotadora',
    scene=dict(
        xaxis=dict(title='FG%'),
        yaxis=dict(title='3P%'),
        zaxis=dict(title='PTS')
    ),
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    )
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='3d-scatter-colorscale')
```

## Tiros Libres
